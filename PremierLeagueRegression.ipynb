```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = df_selected.drop('Rating_x', axis=1)
y = df_selected['Rating_x']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


model = LinearRegression()
model.fit(X_train, y_train)

```

```python
coefficients = pd.Series(model.coef_, index=X.columns)
print(coefficients.sort_values(ascending=False))
```

```python
import numpy as np

# Test seti Ã¼zerindeki tahminler
y_pred = model.predict(X_test)

# MAE
mae = np.mean(np.abs(y_test - y_pred))

# MSE 
mse = np.mean((y_test - y_pred) ** 2)

# RMSE 
rmse = np.sqrt(mse)

# RÂ² Score
ss_total = np.sum((y_test - np.mean(y_test)) ** 2)
ss_res = np.sum((y_test - y_pred) ** 2)
r2_score = 1 - (ss_res / ss_total)

print(f"(MAE): {mae:.4f}")
print(f"(MSE): {mse:.4f}")
print(f"(RMSE): {rmse:.4f}")
print(f"RÂ² Score: {r2_score:.4f}")
```

```python
df['is_star'] = ((df['Rating_x'] >= 7.0) & (df['Rating_y'] > 0.6)).astype(int)

df[['Rating_x', 'Rating_y', 'is_star']].head()
```

```python
class_counts = df['is_star'].value_counts()
total = len(df)

class_percentages = (class_counts / total) * 100

print(class_counts)
print(class_percentages)

```

**It looks like there is imbalance for is_star column. This situation might lead model to predict all players as majority class. (0 for this case.) If this was a malignant-benign project, that would be big problem. We need to prevent that. Accuracy and precision might appear high but misleading. Because model always predict majority class.**

```python
selected_features = [
    'Goals_x', 'Assists_x', 'SpG_x', 'KeyP_x', 'AerialsWon_x',
    'Drb_x', 'Tackles_x', 'Inter_x', 'Blocks_x', 'Rating_y'
]

# X ve y oluÅŸtur
X = df[selected_features].values
y = df['is_star'].values.reshape(-1, 1)

X = np.concatenate([np.ones((X.shape[0], 1)), X], axis=1)

# BaÅŸlangÄ±Ã§ theta deÄŸerleri
theta = np.zeros((X.shape[1], 1))


```

```python
def sigmoid(z): 
    return 1 / (1 + np.exp(-z))

def compute_cost(X, y, theta):
    m = len(y)
    h = sigmoid(X @ theta)
    cost = -(1/m) * np.sum(y * np.log(h) + (1 - y) * np.log(1 - h))
    return cost

def gradient_descent(X, y, theta, alpha, num_iters):
    m = len(y)
    cost_history = []

    for _ in range(num_iters):
        h = sigmoid(X @ theta)
        gradient = (1/m) * (X.T @ (h - y))
        theta -= alpha * gradient
        cost_history.append(compute_cost(X, y, theta))
        
    return theta, cost_history
```

```python
alpha = 0.1
num_iters = 1000
theta, cost_history = gradient_descent(X, y, theta, alpha, num_iters)

def predict(X, theta):
    probs = sigmoid(X @ theta)
    return (probs >= 0.5).astype(int)

y_pred = predict(X, theta)

```

```python
feature_names = ['Intercept'] + selected_features
for name, coef in zip(feature_names, theta.flatten()):
    print(f"{name}: {coef:.4f}")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

accuracy = accuracy_score(y, y_pred)

precision = precision_score(y, y_pred)

recall = recall_score(y, y_pred)

f1 = f1_score(y, y_pred)

cm = confusion_matrix(y, y_pred)

print(f"Accuracy:  {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1 Score:  {f1:.4f}")
print("Confusion Matrix:")
print(cm)
```

**Precision can remain high even if many true stars are missed, as long as the few predicted stars are correct. Recall often suffers because the model tends to predict the majority class (non-stars), missing many actual stars. In this context, F1-score is the most reliable metric because it balances the cost of false positives and false negatives, which is crucial when identifying rare but valuable star players.**
